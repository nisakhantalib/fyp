```python
import os
import pandas as pd
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
from google.colab import files


path_to_C50train = '/content/drive/MyDrive/Colab Notebooks/reuterstrain/C50train' # if using Google Drive
path_to_C50test = '/content/drive/MyDrive/Colab Notebooks/reuterstrain/C50test'


# Function to get approximately 250 words from a text file
def get_approx_250_words(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    #tokenize the text into sentences
    sentences=sent_tokenize(text)
    word_count=0
    selected_text=''

    #loop over the sentences and keep adding them until reach the 250 words limit
    for sentence in sentences:
      words=word_tokenize(sentence)
      word_count+=len(words)
      if word_count > 250:
        break
      else:
        selected_text+= ' ' + sentence

    return selected_text.strip()


# Initialize an empty DataFrame
df = pd.DataFrame(columns=['text', 'label'])

# Loop through each author's folder, get 50 text files, and add to the DataFrame
authors = os.listdir(path_to_C50train)
for author in authors[:50]:
    author_folder = os.path.join(path_to_C50train, author)
    text_files = os.listdir(author_folder)
    for text_file in text_files[:100]:
        file_path = os.path.join(author_folder, text_file)

        text = get_approx_250_words(file_path)

        df = df.append({'label': author, 'text': text}, ignore_index=True)

for author in authors[:50]:
    author_folder = os.path.join(path_to_C50test, author)
    text_files = os.listdir(author_folder)
    for text_file in text_files[:100]:
        file_path = os.path.join(author_folder, text_file)

        text = get_approx_250_words(file_path)

        df = df.append({'label': author, 'text': text}, ignore_index=True)

# Save the DataFrame to an Excel file
df.to_excel('50human100eachwithname.xlsx', index=False)

# Download the file (if you're running this on Colab and want to download the Excel file to your local machine)
files.download('50human100eachwithname.xlsx')
```
